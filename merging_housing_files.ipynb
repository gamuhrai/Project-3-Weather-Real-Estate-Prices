```python
import pandas as pd
```

```python
df1 = pd.read_csv('1b_5yrs_filtered.csv')
df2 = pd.read_csv('2b_5yrs_filtered.csv')
df3 = pd.read_csv('3b_5yrs_filtered.csv')
```

```python
df1['Source'] = '1 Bedroom'
df2['Source'] = '2 Bedroom'
df3['Source'] = '3 Bedroom'
```

```python
combined_df = pd.concat([df1, df2, df3], ignore_index=True)
combined_df = combined_df[['Source'] + [col for col in combined_df.columns if col != 'Source']]
print(combined_df)
```

```python
combined_df.to_csv('combined_housing_data.csv', index=False)
```
