```python
import pandas as pd
import sqlite3
```

```python
# Read CSV into a pandas DataFrame
df = pd.read_csv('/Users/jeanjin/Desktop/DSBC/project_3/1bed_data.csv')

# Create a connection to SQLite database
conn = sqlite3.connect('1bed_data.db')

# Write the DataFrame content into the SQLite database
df.to_sql('1bed_data', conn, if_exists='replace', index=False)

# Close the connection
conn.close()
```
