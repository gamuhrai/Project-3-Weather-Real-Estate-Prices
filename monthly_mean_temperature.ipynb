```python
import pandas as pd
```

```python
temperature_df = pd.read_csv('combined_temperatures.csv', parse_dates=['time'], infer_datetime_format=True)
print(temperature_df)
```

```python
temperature_df['Month'] = temperature_df['time'].dt.month
temperature_df['Year'] = temperature_df['time'].dt.year
print(temperature_df)
```

```python
monthly_mean_temperature_atlanta = temperature_df.groupby(['Year', 'Month'])['temperature_2m_mean_Atlanta'].mean().reset_index()
print(monthly_mean_temperature_atlanta)
monthly_mean_temperature_atlanta.to_csv('monthly_mean_tempature_atlanta.csv')
```
